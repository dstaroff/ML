## Функция одной переменной
<p>Проведем расчет на различных вариантах и посмотрим на график.</p>

```python
coords = list()

coords_min = main(function=f,
                  x0=array([2]),
                  a=0,
                  b=4,
                  lambd=0.1,
                  epsilon=0.2,
                  t='constant')
print('{0}\n\t(X: {1}; Y: {2})'.format('constant', coords_min[0], f(coords_min, 1)))
coords.append((coords_min, 'constant'))

coords_min = main(function=f,
                  x0=array([2]),
                  a=0,
                  b=4,
                  lambd=0.1,
                  epsilon=0.2,
                  t='constant-descent')
print('{0}\n\t(X: {1}; Y: {2})'.format('constant-descent', coords_min[0], f(coords_min, 1)))
coords.append((coords_min, 'constant-descent'))

coords_min = main(function=f,
                  x0=array([2]),
                  a=0,
                  b=4,
                  lambd=0.1,
                  epsilon=0.2,
                  t='optimal-descent')
print('{0}\n\t(X: {1}; Y: {2})'.format('optimal-descent', coords_min[0], f(coords_min, 1)))
coords.append((coords_min, 'optimal-descent'))

coords_min = main(function=f,
                  x0=array([2]),
                  a=0,
                  b=4,
                  lambd=0.1,
                  epsilon=0.2,
                  t='monte-carlo')
print('{0}\n\t(X: {1}; Y: {2})'.format('monte-carlo', coords_min[0], f(coords_min, 1)))
coords.append((coords_min, 'monte-carlo'))

plot(f, coords, (0, 4))
```

<p>Как можно заметить, гибридный градиентный спуск с использованием методики Monte Carlo наиболее точно находит минимум, в отличие от обычных методик.</p>
# Функция Розенброка
<p>Проведем расчет на различных вариантах и посмотрим на графики.</p>

```python
coords = list()

coords_min = main(function=f1,
                  x0=array([1, 2]),
                  a=0,
                  b=4,
                  lambd=0.1,
                  epsilon=200,
                  t='monte-carlo')
print('{0}\n\t(X: {1}; Y: {2}; Z: {3})'.format('monte-carlo', coords_min[0], coords_min[1], f1(coords_min, 2)))
coords.append((coords_min, 'monte-carlo'))

plot(f1, coords, (0, 4))
coords.clear()

coords_min = main(function=f1,
                  x0=array([-20, -20]),
                  a=-16,
                  b=16,
                  lambd=0.1,
                  epsilon=100000,
                  t='monte-carlo')
print('{0}\n\t(X: {1}; Y: {2}; Z: {3})'.format('monte-carlo', coords_min[0], coords_min[1], f1(coords_min, 2)))
coords.append((coords_min, 'monte-carlo'))

plot(f1, coords, (-16, 16))
```
